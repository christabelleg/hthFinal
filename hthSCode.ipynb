```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

smmh_data = pd.read_csv('smmh.csv')
smmh_data
```

```python
#rename columns
new_column_names = {
    'Timestamp': 'timestamp',
    '1. What is your age?': 'age',
    '2. Gender': 'gender',
    '3. Relationship Status': 'relationship_status',
    '4. Occupation Status': 'occupation_status',
    '5. What type of organizations are you affiliated with?': 'affiliated_organizations',
    '6. Do you use social media?': 'social_media_use',
    '7. What social media platforms do you commonly use?': 'social_media_platforms',
    '8. What is the average time you spend on social media every day?': 'avg_time',
    '9. How often do you find yourself using Social media without a specific purpose?': 'no_purpose',
    '10. How often do you get distracted by Social media when you are busy doing something?': 'distracted',
    "11. Do you feel restless if you haven't used Social media in a while?": 'restless',
    '12. On a scale of 1 to 5, how easily distracted are you?': 'distraction_rating',
    '13. On a scale of 1 to 5, how much are you bothered by worries?': 'worries_rating',
    '14. Do you find it difficult to concentrate on things?': 'concentration_rating',
    '15. On a scale of 1-5, how often do you compare yourself to other successful people through the use of social media?': 'comparison_rating',
    '16. Following the previous question, how do you feel about these comparisons, generally speaking?': 'comparison_comments',
    '17. How often do you look to seek validation from features of social media?': 'seekValidation_rating',
    '18. How often do you feel depressed or down?': 'depression_rating',
    '19. On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?': 'interestFluctuation_rating',
    '20. On a scale of 1 to 5, how often do you face issues regarding sleep?': 'sleepIssues_rating',
}

smmh_data = smmh_data.rename(columns=new_column_names)
smmh_data
```

```python
#1. What is the distribution of social media usage among different age groups? 

age_social_media_usage = smmh_data.groupby(['age', 'social_media_use']).size().reset_index(name='count')

# bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=age_social_media_usage, x='age', y='count', hue='social_media_use')
plt.title('Distribution of Social Media Usage Among Different Age Groups')
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.legend(title='Social Media Usage', loc='upper right')
plt.xticks(rotation=90) 

plt.tight_layout()
plt.show()

```

```python
#2. Is there a difference in social media usage between different genders? 

#group non-binary responses together
non_binary_responses = ['NB', 'Non-binary', 'Non binary', 'Nonbinary']
smmh_data['gender_grouped'] = smmh_data['gender'].apply(lambda x: 'Non-binary' if x.strip() in non_binary_responses else x)

social_media_users = smmh_data[smmh_data['social_media_use'] == 'Yes']
gender_social_media_users = social_media_users.groupby('gender_grouped').size().reset_index(name='count')

# Plotting the bar chart for social media users by grouped gender
plt.figure(figsize=(8, 6))
sns.barplot(data=gender_social_media_users, x='gender_grouped', y='count', hue='gender_grouped', legend=False, palette='Set2')
plt.title('Social Media Users by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=90) 
plt.tight_layout()
plt.show()

```

```python
#3. How does relationship status correlate with social media habits and mental health?
# Grouping data by relationship status and calculating mean depression ratings
depression_by_status = smmh_data.groupby('relationship_status')['depression_rating'].mean().reset_index()

# Plotting the pie chart
plt.figure(figsize=(10, 6))
sns.barplot(data=depression_by_status, x='relationship_status', y='depression_rating', hue='relationship_status', legend=False, palette='viridis')
plt.title('Average Depression Rating from Social Media Use by Relationship Status')
plt.xlabel('Relationship Status')
plt.ylabel('Average Depression Rating')
plt.tight_layout()
plt.show()
```

```python
#4. How does frequency of social media usage correlate with feelings of depression?
avg_time_grouped = smmh_data.groupby('avg_time')['depression_rating'].mean().reset_index()

#line plot
plt.figure(figsize=(8, 6))
plt.plot(avg_time_grouped['avg_time'], avg_time_grouped['depression_rating'], marker='o', linestyle='-')
plt.title('Average Time on Social Media vs. Depression Rating')
plt.xlabel('Average Time on Social Media')
plt.ylabel('Depression Rating')
plt.xticks(rotation=90) 
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
#5. What age group is most affected by comparison on social media?
smmh_data['age_group'] = pd.cut(smmh_data['age'], bins=range(15, 101, 10), right=False)

age_comparison = smmh_data.groupby('age_group')['comparison_rating'].mean().reset_index()

#bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=age_comparison, x='age_group', y='comparison_rating', hue= 'age_group', legend=False, palette='viridis')
plt.title('Average Comparison Rating by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Comparison Rating')
plt.tight_layout()
plt.show()

```

```python
#6. How does comparison on social media correlate with feelings of distress?
smmh_data_sorted = smmh_data.sort_values('comparison_rating')

#line graph
plt.figure(figsize=(10, 6))
sns.lineplot(data=smmh_data_sorted, x='comparison_rating', y='depression_rating')
plt.title('Relationship between Comparison in Social Media and Depression Rating')
plt.xlabel('Comparison Rating')
plt.ylabel('Depression Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#7. Who spends most time with social media?
time_mapping = {
    'Less than an Hour': 1,
    'Between 2 and 3 hours': 2.5,
    'Between 3 and 4 hours': 3.5,
    'More than 5 hours': 5
}
smmh_data['avg_time_numeric'] = smmh_data['avg_time'].map(time_mapping)

occupation_avg_time = smmh_data.groupby('occupation_status')['avg_time_numeric'].mean().reset_index()

#bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=occupation_avg_time, x='occupation_status', y='avg_time_numeric', hue='occupation_status', legend=False, palette='Set2')
plt.title('Average Time Spent on Social Media by Occupation Status')
plt.xlabel('Occupation Status')
plt.ylabel('Average Time Spent (h)')
plt.tight_layout()
plt.show()
```

```python
#8. Does occupation status relate to seeking validation on social media?
occupation_validation = smmh_data.groupby('occupation_status')['seekValidation_rating'].mean().reset_index()

#bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=occupation_validation, x='occupation_status', y='seekValidation_rating', hue = 'occupation_status', legend=False,palette='pastel')
plt.title('Average Validation Seeking on Social Media by Occupation Status')
plt.xlabel('Occupation Status')
plt.ylabel('Average Validation Seeking Rating')
plt.tight_layout()
plt.show()
```

```python
#9. Is there a difference in social media's effect on mental health between genders?
non_binary_responses = ['NB', 'Non-binary', 'Non binary', 'Nonbinary']
smmh_data['gender_grouped'] = smmh_data['gender'].apply(lambda x: 'Non-binary' if x.strip() in non_binary_responses else x)

gender_depression_sum = smmh_data.groupby('gender_grouped')['depression_rating'].mean().reset_index()

#pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_depression_sum['depression_rating'], labels=gender_depression_sum['gender_grouped'], autopct='%1.1f%%', startangle=140)
plt.title('Depression Rating by Gender')
plt.axis('equal')
plt.show()
```

```python
#10. Does social media's affect on sleep issues vary among different age groups?
age_sleep_issues = smmh_data.groupby('age')['sleepIssues_rating'].mean().reset_index()

#bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=age_sleep_issues, x='age', y='sleepIssues_rating', hue = 'age', legend=False, palette='coolwarm')
plt.title('Sleep Issues Rating Across Different Age Groups')
plt.xlabel('Age')
plt.ylabel('Sleep Issues Rating')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
